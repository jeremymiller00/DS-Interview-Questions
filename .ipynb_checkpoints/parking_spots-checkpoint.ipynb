```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Parking Spot Selection (interviewqs.com) - Sheet1.csv")
```

```python
selections = pd.DataFrame(data['Spot #'].value_counts())
```

```python
selections.reset_index(inplace=True)
```

```python
selections.rename( {'index': 'spot', 'Spot #': 'count'}, axis=1, inplace=True)
```

```python
selections
```

```python
selections['count'].sum() == data.shape[0]
```

```python
selections['p'] = selections['count'] / data.shape[0]
```

```python
selections.sort_values('spot', inplace=True)
```

```python
selections
```

```python
fig, ax = plt.subplots(figsize=(12,6))
x = selections['spot'].values

# ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1, label='frozen pmf')
# ax.legend(loc='best', frameon=False)

ax.plot(x, selections[selections['spot'] == x]['p'], 'bo', ms=8, label='binom pmf')
ax.vlines(x, 0, selections[selections['spot'] == x]['p'], colors='b', lw=5, alpha=0.5)

plt.show()
```
