Using Python, create a line chart with both the daily and 7-day moving average for the close price for Apple stock. You can download a CSV with prices here.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
aapl = pd.read_csv('../data/AAPL.csv')
```

```python
aapl['Date'] = pd.to_datetime(aapl['Date'])
```

```python
aapl['DailyAvg'] = (aapl['High'] + aapl['Low']) / 2
```

```python
aapl['7DayAvg'] = aapl['DailyAvg'].rolling(7).mean()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(x='Date', y='DailyAvg', data=aapl, ax=ax, label='Daily')
sns.lineplot(x='Date', y='7DayAvg', data=aapl, ax=ax, label='7-Day')
ax.grid()
ax.set_title("Daily and 7 Day Average Apple Stock Price")
fig.show()
```
