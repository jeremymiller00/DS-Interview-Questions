```python
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 10000
dist = expon(scale=mu)
```

```python
x = np.linspace(dist.ppf(0.01), dist.ppf(0.99), 100)
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, dist.pdf(x), 'b', lw=5, alpha=0.6, label='expon pdf')
ax.grid()
ax.fill_between(x[x>3000], dist.pdf(x[x>3000]), alpha=0.2)
ax.vlines(3000, 0, dist.pdf(3000))
plt.show()
```
