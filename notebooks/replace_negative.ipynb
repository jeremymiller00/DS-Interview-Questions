Replacing bad data with Pandas 
Data Analysis Python Pandas Data Cleaning

Suppose you are given a dataframe, df, that contains various negative values. In the context of your work, negative values can be considered 'bad data'. Write a function in Python (using Pandas) that replaces these bad values with the group mean.

Solution will be written in Python for premium users.

```python
import pandas as pd
import numpy as np
```

```python
group_1 = np.random.uniform(-5, 5, 1000)
group_2 = np.random.uniform(-1, 10, 1000)
```

```python
df1 = pd.DataFrame(group_1, columns=['feature'])
df1['group'] = 1
df2 = pd.DataFrame(group_2, columns=['feature'])
df2['group'] = 2
df = pd.concat([df1,df2])
df = df.sample(frac=1)
```

```python
df_pos = df[df['feature'] >= 0]
df_neg = df[df['feature'] < 0]
```

```python
means = df_pos.groupby('group').mean()
mean_1 = means.iloc[0]['feature']
mean_2 = means.iloc[1]['feature']
```

```python
df_neg_1 = df_neg[df_neg['group'] == 1]
df_neg_2 = df_neg[df_neg['group'] == 2]
```

```python
df_neg_1['mean'] = mean_1
df_neg_2['mean'] = mean_2
```

```python
df_neg_1.drop('feature', axis=1, inplace=True)
df_neg_2.drop('feature', axis=1, inplace=True)
```

```python
df_neg_1.rename({'mean': 'feature'}, axis=1, inplace=True)
df_neg_2.rename({'mean': 'feature'}, axis=1, inplace=True)
```

```python
df_neg = pd.concat([df_neg_1, df_neg_2])
```

```python
df = pd.concat([df_pos, df_neg])
df = df.sample(frac=1)
```
