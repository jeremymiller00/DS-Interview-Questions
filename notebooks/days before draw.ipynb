You are drawing from a normally distributed random variable X ~ N(0, 1) once a day. What is the approximate expected number of days until you get a value of more than 2?

```python
from scipy.stats import norm, geom
import matplotlib.pyplot as plt
import numpy as np
```

```python
p = 1 - norm.cdf(2)
```

```python
print(f"For any given day, the probability of getting a value > 2 = {p.round(5)}")
```

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(norm.ppf(0.01),
                norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x),
       'r-', lw=3, alpha=0.6, label='norm pdf')
section = np.arange(2, 2.5, 1/20.)
ax.fill_between(section, norm.pdf(section), alpha=0.2)
ax.grid()
```

We can model days until we get a value greater than 2 as a geometric distribution (failures before a success)

```python
geo_mean = geom.stats(p, moments='m')
```

```python
geo_mean
```

```python
fig, ax = plt.subplots(1, 1)
x = np.arange(geom.ppf(0.01, p),
              geom.ppf(0.99, p))
rv = geom(p)
ax.plot(x, geom.pmf(x, p), 'bo', ms=4, label='geom pmf')
ax.vlines(geo_mean, 0, rv.pmf(geo_mean), colors='r', lw=10, alpha=1)
ax.grid()
```

Answer: 44 days
